```python
import pandas as pd
import numpy as np
import sys
import os

# Get the directory where this notebook is located
notebook_dir = os.path.dirname(os.path.abspath(''))

# Add the PythonPrep directory to the path
pythonprep_dir = os.path.join(os.path.dirname(notebook_dir), 'PythonPrep') if 'PythonPrep' not in notebook_dir else notebook_dir
sys.path.append(pythonprep_dir)

from paths import main_path

import warnings
warnings.filterwarnings("ignore")
```

```python
path_input = main_path + "/Democracy/Democracy_Main/MainAnalysis/input"
leh = pd.read_csv(main_path + '/Democracy/PythonData/healthy-life-expectancy-at-birth.csv')
leh
```

```python
leh.rename(columns={
    'Entity': 'country',
    'Healthy life expectancy (HALE) at birth (years) - Sex: both sexes': 'life_expectancy_hale',
}, inplace=True)

leh = leh.drop('Code', axis=1)
leh
```

```python
df = leh[leh['Year']!=2015]
df
```

```python
df['prev_life_expectancy'] = df.groupby('country')['life_expectancy_hale'].shift(1)
df['growth_rate'] = ((df['life_expectancy_hale'] - df['prev_life_expectancy']) / df['prev_life_expectancy']) * 100

df['growth_column'] = df.groupby('country')['Year'].shift(1).astype(str) + '_' + df['Year'].astype(str)

df_pivot = df.pivot(index='country', columns='growth_column', values='growth_rate').reset_index()
df_pivot = df_pivot.drop('nan_2000', axis=1)

df_pivot.columns.name = None
df_pivot.columns = ['country', 'hale_2000_2010', 'hale_2010_2019']
```

```python
df_pivot
```

```python
df_pivot['hale_2000_2019'] = (df_pivot['hale_2000_2010'] + df_pivot['hale_2010_2019']) / 2
df = df_pivot.drop(['hale_2000_2010', 'hale_2010_2019'], axis=1)
```

```python
result_df = df
```
