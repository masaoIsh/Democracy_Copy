```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import sys
import os

# Get the directory where this notebook is located
current_dir = os.getcwd()
pythonprep_dir = os.path.join(current_dir, '..', '..')
pythonprep_dir = os.path.abspath(pythonprep_dir)
sys.path.append(pythonprep_dir)

from paths import main_path
```

```python
df = pd.read_csv(main_path + '/Democracy/PythonData/population-and-demography.csv')
df


```

```python
df = df[(df['Year'] >= 1960)]
df
```

```python
df.columns
```

```python
df.rename(columns={'Population - Sex: all - Age: all - Variant: estimates': 'Population', 'Entity': 'country'},
          inplace=True)
```

```python
df
```

```python
df['growth_rate'] = df.groupby('country')['Population'].pct_change() * 100

```

```python
periods = {
    '2001-2019': (2001, 2019),
    #'2019-2022': (2019, 2022),
}

result_df = pd.DataFrame()

for period, (start_year, end_year) in periods.items():
    period_df = df[(df['Year'] >= start_year) & (df['Year'] <= end_year)]
    avg_mortality = period_df.groupby('country')['growth_rate'].mean().rename(period)
        
    result_df = pd.concat([result_df, avg_mortality], axis=1)    
        
result_df.reset_index(inplace=True)
result_df['country'] = result_df['index']
result_df = result_df.drop('index', axis=1)
result_df = result_df[['country'] + [col for col in result_df.columns if col != 'country']]
    
result_df.rename(columns={
    '2001-2019': 'avg_pop_growth_2001_2019',
    #'2019-2022': 'avg_pop_growth_2019_2022'

}, inplace=True)
    
result_df
```
