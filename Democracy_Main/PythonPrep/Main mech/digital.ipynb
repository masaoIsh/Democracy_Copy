```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

import sys
import os

# Get the directory where this notebook is located
notebook_dir = os.path.dirname(os.path.abspath(''))

# Add the PythonPrep directory to the path
pythonprep_dir = os.path.join(os.path.dirname(notebook_dir), 'PythonPrep') if 'PythonPrep' not in notebook_dir else notebook_dir
sys.path.append(pythonprep_dir)

from paths import main_path
```

```python
### v2smpolsoc
```

```python
path_input = main_path + "/Democracy/Democracy_Main/MainAnalysis/input"
file_path = f"{path_input}/channels21st/digital-society-project/DigitalSocietyProject-v3-STATA/DSP-Dataset-v3.dta"
variable = 'v2smpolsoc'  
data = pd.read_stata(file_path)
columns_to_keep = ['country_text_id', 'year', variable]
data = data[columns_to_keep]
data = data[data['country_text_id'] != ""]
data_filtered = data[(data['year'] > 2000) & (data['year'] < 2020)]
data_filtered['growth_rate'] = data_filtered.groupby('country_text_id')[variable].pct_change() * 100

result_df = data_filtered.groupby('country_text_id').agg(
    **{f'change_{variable}_2001_2019': (variable, lambda x: x.iloc[-1] - x.iloc[0])}, 
    **{f'mean_{variable}_2001_2019': (variable, 'mean')},  
    **{f'gr_{variable}_2001_2019': ('growth_rate', 'mean')}
).reset_index()

df_2000 = data[data['year'] == 2000]
pivot_df = df_2000.pivot_table(index='country_text_id', columns='year', values=variable).reset_index()
df_renamed = pivot_df.rename(columns={2000: f'con_{variable}_2000'})
merged_df = pd.merge(result_df, df_renamed, on='country_text_id', how='left')
result_df1 = merged_df.dropna()
```
