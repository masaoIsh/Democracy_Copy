```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import sys
import os

# Get the directory where this notebook is located
notebook_dir = os.path.dirname(os.path.abspath(''))

# Add the PythonPrep directory to the path
pythonprep_dir = os.path.join(os.path.dirname(notebook_dir), 'PythonPrep') if 'PythonPrep' not in notebook_dir else notebook_dir
sys.path.append(pythonprep_dir)

from paths import main_path
```

```python
file_path = main_path + '/Democracy/Democracy_Main/MainAnalysis/input/outcomes/happiness/DataForTable2.1WHR2023.xls'

df = pd.read_excel(
        file_path,
        sheet_name="Sheet1",
        skiprows=0,
        engine="xlrd"          # explicitly tell pandas to use xlrd
)
df
```

```python
df['growth_rate_life_ladder'] = df.groupby('Country name')['Life Ladder'].pct_change() * 100
df['growth_rate_positive_affect'] = df.groupby('Country name')['Positive affect'].pct_change() * 100
```

```python
periods = {
    '2010_2019': (2010, 2019),
    '2020_2022': (2020, 2022),
}

result_df = pd.DataFrame()

for period, (start_year, end_year) in periods.items():
    period_df = df[(df['year'] >= start_year) & (df['year'] <= end_year)]
    avg_life_ladder = period_df.groupby('Country name')['growth_rate_life_ladder'].mean().rename(f'life_ladder_{period}')
    avg_positive_affect = period_df.groupby('Country name')['growth_rate_positive_affect'].mean().rename(f'positive_affect_{period}')
    
    result_df = pd.concat([result_df, avg_life_ladder, avg_positive_affect], axis=1)

result_df.reset_index(inplace=True)
result_df['country'] = result_df['index']
result_df = result_df.drop('index', axis=1)
result_df = result_df[['country'] + [col for col in result_df.columns if col != 'country']]

# result_df.rename(columns={
#     '2011-2020': 'life_expectancy_2011_2020',
#     '2020-2022': 'life_expectancy_2020_2022',
#     '2011-2020': 'life_expectancy_2011_2020',
#     '2020-2022': 'life_expectancy_2020_2022',
# }, inplace=True)

result_df
```
