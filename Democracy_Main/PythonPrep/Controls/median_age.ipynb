```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import sys
import os

# Get the directory where this notebook is located
notebook_dir = os.path.dirname(os.path.abspath(''))

# Add the PythonPrep directory to the path
pythonprep_dir = os.path.join(os.path.dirname(notebook_dir), 'PythonPrep') if 'PythonPrep' not in notebook_dir else notebook_dir
sys.path.append(pythonprep_dir)

from paths import main_path
```

```python
df = pd.read_csv(main_path + '/Democracy/PythonData/median-age.csv')
df

```

```python
df = df[(df['Year'] >= 1999) & (df['Year'] <= 2020)]
df = df.drop('Median age - Sex: all - Age: all - Variant: medium', axis=1)
df
```

```python
pivot_table = df.pivot_table(index='Entity', columns='Year', 
                             values='Median age - Sex: all - Age: all - Variant: estimates')

pivot_table.columns = ['medianage_' + str(col) for col in pivot_table.columns]
pivot_table = pivot_table.reset_index()
pivot_table = pivot_table.rename(columns={
    'Entity': 'country'
})

pivot_table
```

```python
pivot_table.to_csv(main_path + '/Democracy/Democracy_Main/MainAnalysis/input/controls/median_age/medianage.csv')
```
